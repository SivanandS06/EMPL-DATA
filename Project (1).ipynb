```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
empl_data=pd.read_csv("myexcel - myexcel.csv.csv")
empl_data
```

#### PREPROCESSING : REPLACING INCORRECT "HEIGHT" DATA

```python
empl_data['Height']=np.random.uniform(150,180,size=len(empl_data))
empl_data
```

##### 1. How many are there in each team and the percentage splitting with respect to the total employees

```python
team_counts=empl_data['Team'].value_counts()
team_percentile=(team_counts / len(empl_data))*100
summary_team=pd.DataFrame({'Count':team_counts,'Percentage':team_percentile})
summary_team
```

##### 2. Segregate the Employees W.R.T different positions

```python
positional_count=empl_data ['Position'].value_counts()
positional_count
```

#### 3. Find from which age group most of the employees belong to

```python
empl_data['Age'].value_counts()
```

##### 4. Finding out under which team and position spending in terms of salary is high

```python
spending_data=empl_data[['Team','Position','Salary']]
spending_data
```

##### 5. Finding if there is any correlation between age and salary

```python
sns.scatterplot(x="Age",y="Salary",data=empl_data)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title("Correlation Between Age and Salary")
plt.show()
```
